```python
import pandas as pd

# Load the dataset
file_path = "../data/Game_of_Thrones_Script.csv"
data = pd.read_csv(file_path)

# Standardize character names
data['Name'] = data['Name'].str.title()

data.head()
```

```python
# Get number of lines per character and organize them largest to smallest
lines_per_character = data['Name'].value_counts().reset_index()
lines_per_character.columns = ['Character', 'Line Count']

# Find the proportion of lines of each character in regards to the top speaker
top_speaker_lines = lines_per_character.iloc[0]['Line Count']
lines_per_character['Proportion to Top Speaker'] = lines_per_character['Line Count'] / top_speaker_lines

# Display the results
lines_per_character.head(20)
```
